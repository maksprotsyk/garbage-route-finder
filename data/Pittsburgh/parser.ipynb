```python
import pandas as pd

def create_containers(filename, container_type):
    df = pd.read_csv(filename)
    df.head()

    containers = []
    for row in df.iterrows():
        latitude = row[1]["y"]
        longitude = row[1]["x"]
        containers.append(
            {
                "type": "RegularContainer",
                "latitude": float(latitude),
                "longitude": float(longitude),
                "start": 0,
                "finish": 100000000
            }
        )    
    return containers

files = [
    "pitsburg.csv"
]

containers = []
types = ["mixed"]
for file in files:
    type_name = "Regular"
    containers.extend(create_containers(file, f"{type_name}Container"))
    types.append(type_name)
containers
```

```python
import random
random.seed(0)

def get_random_location(containers):
    index = random.randrange(0, len(containers) - 1)
    res = containers[index]["latitude"], containers[index]["longitude"]
    del containers[index]
    return res
trucks_location = get_random_location(containers)
landfill_location = (40.505002,-79.776352)
```

```python
import json
truck_types = [
    {
      "type": "SmallTruck",
      "speed": 150,
      "fuel_consumption": 7,
      "fuel_capacity": 150,
      "capacities": {"mixed": 100},
      "unloading_time": 15
    },
    {
      "type": "BigTruck",
      "speed": 10,
      "fuel_consumption": 0.5,
      "fuel_capacity": 50000,
      "capacities": {"mixed": 200},
      "unloading_time": 15
    },
]


container_types = [
    {
        "capacity": 10,
        "processing_time": 10,
        "garbage_type": "mixed",
        "type": "RegularContainer"
    }
    
]

trucks = []
for i in range(10):
        trucks.append(
            {
                "type": "BigTruck",
                "latitude": trucks_location[0],
                "longitude": trucks_location[1],
                "start": 0,
                "finish": 100000000
            }
        )
landfills = [
    {
        "latitude": landfill_location[0],
        "longitude": landfill_location[1],
        "start": 0,
        "finish": 100000000,
        "max_trucks": 100
    }
]
data = {"containers": containers, "description": "Netherland", "trucks": trucks, "landfills": landfills}
json_string = json.dumps(data, indent=4)
with open("input.json", "w") as f:
    f.write(json_string)
```

```python
config_data = {
    "garbage_types": types,
    "container_types": container_types,
    "truck_types": truck_types
}
json_string = json.dumps(config_data, indent=4)
with open("config.json", "w") as f:
    f.write(json_string)
```
